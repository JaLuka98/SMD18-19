## b)
Um das Problem zu beheben, wird der Nenner mithilfe der gegebenen Hinweise umgeschrieben und es ergibt sich die Gleichung
$$\frac{d\sigma}{d\Omega}=\frac{\alpha^2}{s}\left(\frac{2+\sin^2\left(\theta\right)}{\sin^2\left(\theta\right)+\frac{1}{\gamma^2}\cos^2\left(\theta\right)}\right)$$

## c) 
Es ergeben sich in einem Bereich um $\theta = 0$ die folgenden Graphen.


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python


alpha = 7.2973525664*10**(-3)
E = 50*10**9
m = 511*10**3
s = 2*E
gamma = E/m
beta = np.sqrt(1-gamma**(-2))

def f(x):
    return alpha**2/s*((2+np.sin(x)**2)/(1-beta**2*np.cos(x)**2))

def g(x):
    return alpha**2/s*((2+np.sin(x)**2)/(np.sin(x)**2+1/gamma**2*np.cos(x)**2))

x = np.linspace(-0.00000005,0.00000005)

plt.plot(x, f(x),'r-', label='Funktionswerte')
plt.xlabel('x')
plt.ylabel('do/domega')
plt.title('Ursprüngliche Form')
plt.legend
plt.show()
plt.clf()

plt.plot(x, g(x),'r-', label='Funktionswerte')
plt.xlabel('x')
plt.ylabel('do/domega')
plt.title('Verbesserte Form')
plt.legend
plt.show()
    
```

Trägt man diese Konditionszahl gegen $\theta$ in einem Bereich zwischen 0 und $\pi$ auf, ergibt sich der folgende Graph

```python
def k(x):
    return alpha**2/s*(-1*(2*x*np.cos(x)*(1-beta**2*np.cos(x)**2)*np.sin(x)*(beta**2*np.sin(x)**2+beta**2*np.cos(x)**2+2*beta**2-1))/((beta**2*np.cos(x)**2-1)**2*(np.sin(x)**2+2)))

x = np.linspace(0, np.pi)

plt.plot(x, k(x),'r-', label='Funktionswerte')
plt.xlabel('x')
plt.ylabel('k(x)')
plt.legend
plt.show()
```
