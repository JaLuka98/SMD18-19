```python
import numpy as np
import matplotlib.pyplot as plt
```

```python


alpha = 7.2973525664*10**(-3)
E = 50*10**9
m = 511*10**3
s = 2*E
gamma = E/m
beta = np.sqrt(1-gamma**(-2))

def f(x):
    return alpha**2/s*((2+np.sin(x)**2)/(1-beta**2*np.cos(x)**2))

def g(x):
    return alpha**2/s*((2+np.sin(x)**2)/(np.sin(x)**2+1/gamma**2*np.cos(x)**2))

x = np.linspace(-0.00000005,0.00000005)

plt.plot(x, f(x),'r-', label='Funktionswerte')
plt.xlabel('x')
plt.ylabel('do/domega')
plt.legend
plt.show()
plt.clf()

plt.plot(x, g(x),'r-', label='Funktionswerte')
plt.xlabel('x')
plt.ylabel('do/domega')
plt.legend
plt.show()
    
```
