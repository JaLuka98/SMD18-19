```python
import numpy as np
from numpy import random
import matplotlib.pyplot as plt
from scipy import stats
```

```python
# Generates the next number of the sequence generated by a linear congruental generator
def lcg(x, a, b, m):
    return (a*x + b) % m
```

```python
# Setting the params of the lcg used here
b = 3
m = 1024
a_max = 30 # HIER SOLLTEST DU NOCH ETWAS WEITER GEHEN, 30 IST NUR EIN TEST
periodenlängen = []; # create an empty list. Periodenlängen contains the period length for each a
for a in range(a_max):
    #print('Jetzt a = ', a)
    x_i = 0 # We set x_0 = 0 for this part of the exercise
    producedNumbers = []; # In producedNumbers we story the so far generated numbers to compare them. If a number is generated
    # twice we have found the period length
    for i in range(2000):
        if x_i not in producedNumbers:
            producedNumbers.append(x_i)
            #print('Produzierte Zahlen:', producedNumbers)
        else:
            #print('periodenlänge ist', i + 1)
            periodenlängen.append(i);
            #print('Bisherige Periodenlängen', periodenlängen)
            break
        #print('x_i ist ', x_i)
        x_i = lcg(x_i, a, b, m) # Generate the next number

# AN JANNIS: EVENTUELL KANNST DU ES SCHLAUER MACHEN UND STATTDESSEN ABFRAGEN; WANN DAS NÄCHSTE MAL DIE NULL KOMMT.
# DAS SOLLTE EIGENTLICH AUCH KLAPPEN; WIR FANGEN JA MIT 0 AN UND IRGENDWANN SOLLTE DIESE WIEDERKOMMEN!
```

```python
periodenlängen = np.asarray(periodenlängen, dtype=np.float32) # Convert the list to a numpy array
```

```python
alinspace = np.linspace(0, a_max-1, a_max) # We want to plot period length L against the paramter a
plt.plot(alinspace, periodenlängen, 'b-', label='lcg')
plt.grid()
plt.legend()
plt.xlabel('a')
plt.ylabel('Periodenlänge')
plt.show()

# DAS BEABSICHTIGTE VERHALTEN!
# ES SCHLÄGT AUS FÜR a = 1, a = 5, a = 9, ... . DENN DANN IST a-1, ALSO 0,4,8 durch 2 und 4 teilbar!!!!!!!!!!!
```

```python
plt.clf()
a = 1601
b = 3456
m = 10000
```

```python
pValues = []
for x_i in range(0,10000,1000):
    histArray = [x_i/m]
    print(x_i)
    for i in range(10000):
        x_i = lcg(x_i, a, b, m)
        histArray.append(x_i/m)
    #print(histArray)
    binnedData = np.histogram(histArray, bins=10, range=(0,1))[0] # This bins the data in an array
    #print(binnedData)
    pValue = stats.chisquare(binnedData)[1] # Calculates p-value of chi-squared-test. p-value close to 1 => probably uniform
    print(pValue)
    pValues.append(pValue)
    print(pValues)
#plt.hist(histArray, bins=10, range=(0,10000),
#         histtype='bar', label='uniform hist', color='r')
plt.plot(np.linspace(0,10000,10), pValues, label='pValues')
plt.grid()
plt.legend()
plt.xlabel('x_0')
plt.ylabel('p')
plt.show()
```
