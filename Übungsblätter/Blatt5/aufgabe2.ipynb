```python
from sklearn import datasets
from sklearn.decomposition import PCA 
import matplotlib.pyplot as plt
import numpy as np
```

```python
x,y = datasets.make_blobs(n_samples=1000, centers=2,
                                n_features=4, random_state=0)
x = x - x.mean(axis=0) # Zentrierung
plt.scatter(x[:, 0], x[:, 1], marker='o', c=y,
            s=25, edgecolor='k')
plt.xlabel(r'$x$')
plt.ylabel(r'$y$')
plt.title('Darstellung der $x_1$- und $x_2$-Projektion des vierdimensionalen Blobs')
plt.grid()
plt.show()
plt.clf()
```

```python
c = np.cov(x, rowvar=False)
eigw = np.linalg.eigvals(c)
print(eigw)
    
pca = PCA(n_components = 4)
pca.fit(x)
x = pca.fit_transform(x)
print(x)
print(y)

```

```python
x1, x2, x3, x4 = zip(*x)

x1MembersFirst = x1*y
x1MembersFirst = x1MembersFirst[x1MembersFirst != 0]
x1MembersSecond = x1*(((y+1)-2)*(-1))
x1MembersSecond = x1MembersSecond[x1MembersSecond != 0]

x2MembersFirst = x2*y
x2MembersFirst = x2MembersFirst[x2MembersFirst != 0]
x2MembersSecond = x2*(((y+1)-2)*(-1))
x2MembersSecond = x2MembersSecond[x2MembersSecond != 0]

x3MembersFirst = x3*y
x3MembersFirst = x3MembersFirst[x3MembersFirst != 0]
x3MembersSecond = x3*(((y+1)-2)*(-1))
x3MembersSecond = x3MembersSecond[x1MembersSecond != 0]

x4MembersFirst = x4*y
x4MembersFirst = x4MembersFirst[x4MembersFirst != 0]
x4MembersSecond = x4*(((y+1)-2)*(-1))
x4MembersSecond = x4MembersSecond[x4MembersSecond != 0]

plt.subplot(2,2,1)
plt.hist(x1MembersFirst, normed=True, bins = 12, histtype='step', color='black')
plt.hist(x1MembersSecond, normed=True, bins = 12, histtype='step', color='y')
plt.xlabel(r'$x_{1}^{\prime}$')
plt.ylabel('normed counts')
plt.grid()

plt.subplot(2,2,2)
plt.hist(x2MembersFirst, normed=True, bins = 12, histtype='step', color='black')
plt.hist(x2MembersSecond, normed=True, bins = 12, histtype='step', color='y')
plt.xlabel(r'$x_{2}^{\prime}$')
plt.ylabel('normed counts')
plt.grid()

plt.subplot(2,2,3)
plt.hist(x3MembersFirst, normed=True, bins = 12, histtype='step', color='black')
plt.hist(x3MembersSecond, normed=True, bins = 12, histtype='step', color='y')
plt.xlabel(r'$x_{3}^{\prime}$')
plt.ylabel('normed counts')
plt.grid()

plt.subplot(2,2,4)
plt.hist(x4MembersFirst, normed=True, bins = 12, histtype='step', color='black')
plt.hist(x4MembersSecond, normed=True, bins = 12, histtype='step', color='y')
plt.xlabel(r'$x_{4}^{\prime}$')
plt.ylabel('normed counts')
plt.grid()

plt.tight_layout()
plt.show()
plt.clf()
```
