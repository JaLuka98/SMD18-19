```python
import numpy as np
import matplotlib.pyplot as plt


def cdfinv(z, b):
    return b**z

def f(x, b):
    return 1/(np.log(b)*x)
```

```python
b = 5
xlin = np.linspace(1, b, 1000)
sample = np.random.uniform(low = 0.0, high = 1.0, size = 100000)
transformedSample = cdfinv(sample, b)

plt.plot(xlin, f(xlin, b))
plt.hist(transformedSample, normed = True, bins = 50)
plt.grid()
plt.show()
plt.clf()
```

```python
def cdfinv(z, b):
    return 1/(np.pi*b)*np.arccos(1-2*z)

def f(x, b):
    return np.pi*b/2*np.sin(np.pi*b*x)
```

```python
b = 0.5
xlin = np.linspace(0, 1/b, 1000)
sample = np.random.uniform(low = 0.0, high = 1.0, size = 10000)
transformedSample = cdfinv(sample, b)

plt.plot(xlin, f(xlin, b))
plt.hist(transformedSample, normed = True, bins = 100)
plt.grid()
plt.show()
plt.clf()
```
