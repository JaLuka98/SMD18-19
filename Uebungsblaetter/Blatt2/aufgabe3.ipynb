## Teilaufgabe b)
Zu zeigen ist, dass Kurven konstanter Wahrscheinlichkeitsdichte Ellipsen sind.
Die Wahrscheinlichkeitsdichtefunktion der bivariaten Normalverteilung ist durch
$$
f(x,y) =
      \frac{1}{2 \pi  \sigma_x \sigma_y \sqrt{1-\rho^2}}
      \exp\left(
        -\frac{1}{2(1-\rho^2)}\left[
          \frac{(x-\mu_x)^2}{\sigma_x^2} +
          \frac{(y-\mu_y)^2}{\sigma_y^2} -
          \frac{2\rho(x-\mu_x)(y-\mu_y)}{\sigma_x \sigma_y}
        \right]
      \right)
$$
gegeben.
Soll nun f(x,y) konstant sein, so muss der Exponent konstant sein. Dies ergibt
$$-\frac{1}{(1-\rho^2)}\left[\frac{(x-\mu_x)^2}{\sigma_x^2} + \frac{(y-\mu_y)^2}{\sigma_y^2} - \frac{2\rho(x-\mu_x)(y-\mu_y)}{\sigma_x \sigma_y}\right] = \text{const.}\,.
$$
Das ist eine Ellipsengleichung, da stets durch die Konstante dividiert werden kann. Der Mittelpunkt dieser Ellipse ist $(\mu_x,\mu_y)^\text{T}$ und die quadrierten Halbachsen in $x,y$-Richtung sind $(1-\rho^2) \sigma_{x,y}^2$, falls die Konstante gleich eins gesetzt wird (Minus wird absorbiert). Der Mischterm verdeutlicht, dass die Ellipse gegen die kartesischen Koordinatenachsen verdreht ist.

## Teilaufgabe d)
Gesucht ist eine Transformation durch eine Matrix $M$, welche die alten Variablen $x$ und $y$ auf neue Variablen $x'$ und $y'$ transformiert, welche unkorreliert sein sollen. Die Transformation unserer Wahl lautet
$$
\begin{pmatrix}
           x' \\
           y'
         \end{pmatrix} = 
\begin{pmatrix} 
\cos \alpha & - \sin \alpha \\
\sin \alpha & \cos \alpha 
\end{pmatrix}
\begin{pmatrix}
           x \\
           y
        \end{pmatrix}\,.
$$
Dies stellt eine orthogonale Transformation dar, sodass die Rücktransformation durch
$$
\begin{pmatrix}
           x \\
           y
         \end{pmatrix} = 
\begin{pmatrix} 
\cos \alpha & \sin \alpha \\
- \sin \alpha & \cos \alpha 
\end{pmatrix}
\begin{pmatrix}
           x' \\
           y'
        \end{pmatrix}
$$
gegeben ist. Wir berücksichtigen, dass es später rechnerisch leichter ist, wenn wir vorher eine Verschiebung durchführen, sodass der Mittelpunkt der Ellipse im Koordinatenursprung liegt. Dies beeinflusst die Länge der Halbachsen und den Winkel $\alpha$ nicht. Etwas ungenau bezeichnen wir die Achsen in diesem System auch als $x'$ und $y'$. Da wir aus der Gleichung aber nur Informationen über translatorisch invariante Größen wie die Länge der Halbachsen und den Winkel ziehen, sollte das gerade noch okay sein. Die Verschiebung sorgt dafür, dass in der Ellipsengleichung die $\mu_x$ und $\mu_y$ wegfallen, sodass sich die Ellipsengleichung in gestrichenen Koordinaten als
$$-\frac{1}{(1-\rho^2)}\left[\frac{(x'\cos\alpha-y'sin\alpha)^2}{\sigma_x^2} + \frac{(x'\sin\alpha+y'\cos\alpha)^2}{\sigma_y^2} - \\ \frac{2\rho(x'\cos\alpha-y'\sin\alpha)(x'\sin\alpha + y'\cos\alpha)}{\sigma_x \sigma_y}\right] = \text{const.}$$
schreiben lässt. Ausmultiplizieren und sortieren ergibt 
$$
-\frac{1}{(1-\rho^2)}
\left[
x'^2 \left(\frac{\cos^2\alpha}{\sigma_x^2} - \frac{2\rho \sin\alpha \cos\alpha}{\sigma_x \sigma_y} + \frac{\sin^2\alpha}{\sigma_y^2} \right) + 
y'^2 \left(\frac{\sin^2\alpha}{\sigma_x^2} - \frac{2\rho \sin\alpha \cos\alpha}{\sigma_x \sigma_y} + \frac{\cos^2\alpha}{\sigma_y^2} \right) + \\
2 x' y' \left( \frac{\rho\cos^2\alpha}{\sigma_x\sigma_y} - \frac{\rho \cos^2\alpha}{\sigma_x\sigma_y} - \frac{\sin\alpha \cos\alpha}{\sigma_x^2} + \frac{\sin\alpha\cos\alpha}{\sigma_y^2} \right) 
\right]
= \text{const.}\,.
$$
Wir identifizieren den Koeffizienten vor dem $x'$ als reziprokes Quadrat der ersten Halbachse, analog mit dem $y'$. Explizit hingeschrieben bedeutet das für die Halbachsen $a_{x{,}y}$
$$\begin{align*}
a_x^2 &= (1-\rho^2) \left(\frac{\cos^2\alpha}{\sigma_x^2} - \frac{2\rho \sin\alpha \cos\alpha}{\sigma_x \sigma_y} + \frac{\sin^2\alpha}{\sigma_y^2} \right)^{-1} \,, \\
a_y^2 &= (1-\rho^2) \left(\frac{\sin^2\alpha}{\sigma_x^2} - \frac{2\rho \sin\alpha \cos\alpha}{\sigma_x \sigma_y} + \frac{\cos^2\alpha}{\sigma_y^2} \right)^{-1}\,. \\
\end{align*}$$
Da wir nun eine Ellipse haben sollten, die mit den Koordinatenachsen orientiert ist, sollte der gemischte Term in der Ellipsengleichung verschwinden. Aus dieser Bedingung folgt also
$$
\begin{align*}
\left( \frac{\rho\cos^2\alpha}{\sigma_x\sigma_y} - \frac{\rho \cos^2\alpha}{\sigma_x\sigma_y} - \frac{\sin\alpha \cos\alpha}{\sigma_x^2} + \frac{\sin\alpha\cos\alpha}{\sigma_y^2} \right) = 0 \iff \\
\tan\alpha - \cot\alpha = - \frac{\sigma_x^2-\sigma_y^2}{\rho \sigma_x \sigma_y} \iff \\
\frac{1-\tan^2\alpha}{\tan\alpha} = - \frac{\sigma_x^2-\sigma_y^2}{\rho \sigma_x \sigma_y} \iff \\
\frac{2\tan\alpha}{1-\tan^2\alpha} = - \frac{2\rho \sigma_x \sigma_y}{\sigma_x^2-\sigma_y^2} \iff
\tan(2\alpha) = - \frac{2\rho \sigma_x \sigma_y}{\sigma_x^2-\sigma_y^2} \iff\\
\alpha = \frac{1}{2} \text{arctan}\left(- \frac{2\rho \sigma_x \sigma_y}{\sigma_x^2-\sigma_y^2}\right)\,.
\end{align*}$$
Damit konnte explizit die Formel für $\alpha$ gezeigt werden. Die gestrichenen Standardabweichungen berechnen wir im Code durch eine Drehung der Kovarianzmatrix.

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Ellipse

sigmax = 3.5
sigmay = 1.5
covariance = 4.2
rho = 0.8
cov = np.array([[12.25, 4.2], [4.2, 2.25]])
mu = np.array([4,2])
mux = 4
muy = 2
B = 1/(1-rho**2) * np.array([[sigmax**(-2), -rho/(sigmax*sigmay)], 
                             [-rho/(sigmax*sigmay), sigmay**(-2)]])
alpha = np.degrees(0.5 * np.arctan(2*rho*sigmax*sigmay/(sigmax**2-sigmay**2)))
print(alpha)
print(rho)

ells = [Ellipse(xy=(4,2), width=2*np.sqrt(sigmax**2*(1-rho**2)),
                height=2*np.sqrt(sigmay**2*(1-rho**2)), angle = alpha, 
                fill=False)]

fig = plt.figure(0)
ax = fig.add_subplot(111, aspect='equal')
for e in ells:
    ax.add_artist(e)
    e.set_clip_box(ax.bbox)
ax.set_xlim(0, 8)
ax.set_ylim(0, 4)

def f(x,y):
    k = 1/(2*np.pi*sigmax*sigmay*np.sqrt(1-rho**2))
    Vorfaktor = -1/(2*(1-rho**2))
    return k * np.exp(Vorfaktor*((x-4)**2/sigmax**2 - 
                                 2*rho*(x-4)*(y-2)/(sigmax*sigmay)+ 
                                 (y-2)**2/sigmay**2))
x = np.arange(0,8.0,0.01)
y = np.arange(0.0,4.0,0.01)

X,Y = np.meshgrid(x, y)
Z =f(X,Y)
plt.pcolormesh(X,Y,Z,cmap='inferno')
plt.colorbar()

plt.errorbar(4, 2, yerr=sigmay, xerr=sigmax, ecolor='k',capsize=3)

#Berechnung von sigma_x' und sigma_y'-------------
alpha1 = np.radians(alpha)
M = [[np.cos(alpha1), -np.sin(alpha1)], [np.sin(alpha1), np.cos(alpha1)]]
C = [[sigmax**2, covariance], [covariance, sigmay**2]] 
MTC = np.matmul(np.transpose(M), C)
G = np.matmul(MTC, M)

sigmax1 = np.sqrt(G[0,0])
sigmay1 = np.sqrt(G[1,1])
print(sigmax1)
print(sigmay1)
#----------------------------------------------------

#Neue Achsen und sigmas einzeichnen---------------------
a = -np.tan(alpha)
b = 2-(a*4)
def f(x):
    return a*x+b

x = np.linspace(4-sigmay1*np.sin(alpha1), 4+sigmay1*np.sin(alpha1))
plt.plot(x, f(x), 'r-', label=r'$\sigma_x´/\sigma_y´$') #sigma_y'
x = np.linspace(3.65,4)
plt.plot(x, f(x), 'k--', label='Halbachsen') #neue y-Achse

a = 1/np.tan(alpha) 
b = 2-(a*4)
x = np.linspace(4-sigmax1*np.cos(alpha1), 4+sigmax1*np.cos(alpha1))
plt.plot(x, f(x), 'r-') #sigma_x'
x = np.linspace(4,5.9)
plt.plot(x,f(x), 'k--') #neue x-Achse
#--------------------------------

#Bedingte Erwartungswerte-----------
def E(x):
    return (24*x+44)/70
def G(y):
    return (30*x-8)/56 #dies ist die Umkehrfunktion der oben stehenden Funktion
    
x = np.linspace(0,8)

plt.plot(x, E(x), 'g-', label='E(y|x)')
plt.plot(x, G(y), 'b-', label='E(x|y)')
#-------------------------------------

plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()






```
