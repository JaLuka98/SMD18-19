## Teilaufgabe a)

Der Fluss der kosmischen Strahlung soll konstant im Messzeitraum sein.
Die Zahl der gemessenen Protonen kann mit einer Poissonverteilung modelliert werden.
Daher ergibt sich die Maximum-Likelihood-Funktion als:

\begin{align*}
L(\lambda) = \prod_{i} P(X = x_i)
\end{align*}

wobei wie oben beschrieben:

\begin{align*}
P(X) = \frac{\lambda^x}{x!}exp(-\lambda)
\end{align*}

Es ergibt sich

\begin{align*}
L(\lambda) &= \frac{\lambda^{4135}}{4135!}exp(-\lambda)\frac{\lambda^{4202}}{4202!}exp(-\lambda)\frac{\lambda^{4203}}{4203!}exp(-\lambda)\frac{\lambda^{4218}}{4218!}exp(-\lambda)\frac{\lambda^{4227}}{4227!}exp(-\lambda)
\frac{\lambda^{4231}}{4231!}exp(-\lambda)\frac{\lambda^{4310}}{4310!}exp(-\lambda) \\ &= \frac{\lambda^{29526}}{4135!\cdot4202!\cdot4203!\cdot4218!\cdot4227!\cdot4231!\cdot4310!}exp(-7\lambda)
\end{align*}

Logarithmieren:

\begin{align*}
l(\lambda) &= ln(L(\lambda)) = 29526\cdot ln(\lambda) - 7\lambda - ln(4135!\cdot4202!\cdot4203!\cdot4218!\cdot4227!\cdot4231!\cdot4310!) \\
&\approx 29526\cdot ln(\lambda) - 7\lambda - 216968,51
\end{align*}

Ableiten: 

\begin{align*}
\frac{l(\lambda)}{d\lambda} = 29526\cdot \frac{1}{\lambda} - 7
\end{align*}

Die Ableitung hat eine Nullstelle bei 

\begin{align*}
\lambda = 4218
\end{align*}

Die wahrscheinlichste Z채hlrate ist also $N = 4218$.

Es ergibt sich ein log-Likelihood-Wert von 

\begin{align*}
l(4218) = -37,55
\end{align*}

Und damit ein Likelihood-Wert von 

\begin{align*}
L(4218) = e^{-37,55} = 4,92\cdot 10^{-17}
\end{align*}
